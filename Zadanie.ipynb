```python
import pandas as pd
```

```python
prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),

]
```

```python
df = pd.DataFrame(prices)
df
```

```python
df = pd.DataFrame(prices, columns=["month", "pricePLN"])
df = df.set_index("month")
df
```

```python
df['priceUSD'] = df['pricePLN'] / 4
df
```

```python
df['priceUSD'].plot.line(title='Price of good (USD)', color='red', linestyle='dashed', ylim=(0,2), grid=True)
```
